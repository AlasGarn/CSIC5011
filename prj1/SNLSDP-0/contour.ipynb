```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("matrix_1.dat") as f:
    data1 = np.loadtxt(f, dtype ="float64", delimiter = ",")
with open("matrix_0.dat") as f:
    data0 = np.loadtxt(f, dtype ="float64", delimiter = ",")
```

```python
data = data1 + data0 / 2
```

```python
xlist = range(data.shape[0])
ylist = np.linspace(11, 21, 15)
X, Y = np.meshgrid(xlist, ylist)

#fig, ax = plt.subplots(figsize = (525))

cp = plt.contourf(X, Y, data.T, cmap="rainbow")
plt.colorbar(cp)
plt.ylim(11,21)
plt.title('RMSD of the Structures Recovered by SNLSDP')
plt.ylabel('Contact cutoff, $\AA$')
plt.xlabel("Noise Standard Deviation, $\AA$")
```
