```python
from matplotlib.pyplot import *
import pandas
import numpy as np
%matplotlib inline
from scipy.linalg import svd
from sklearn.decomposition import PCA
from scipy.spatial import KDTree
import matplotlib.image as mpimg
```

# I PCA

```python
data = pandas.read_csv("./data/train.3.txt", header = None)
data_default = data.as_matrix()

data = (data_default - data_default.mean(axis=0))# / data.std(axis=0)
```

# SVD and eigenvalues analysis

```python
k = 64 

U, s, VT = svd(data, full_matrices=False)
S = np.diag(s)

VTk = VT[:k]
Uk = U[:,:k]
Sk = S[:k,:k]
var = s / sum(s)
_ = plot(var[:k])
_ = title("Top " + str(k) + " normalized eigenvalues")
```

# visualizing top-k left vectors

This is a bit confusing, as left singular vectors have the dimensions 1xN, so visualizing them with imshow is kind of unconvenient.

SVD of the covariance matrix however gives us an 1xp vectors, which are identical to right singular in this case

Only right vectors of the SVD of the data are meaningful to visualize

```python
x,y,z = svd(np.cov(data.T), full_matrices=False)

N = 15 # aka k
f, ax = subplots(1,N,figsize=(N*1,2))
for i in range(N):
    ax[i].imshow(x[:,i].reshape((16,16)),interpolation='none', cmap = "Greys")
    ax[i].axis("off")
```

# sort data by the first right vector

```python
N = 25 # number of images to display
vector = 0 # sort by which PC?
projections = np.dot(data, VT.T)[:,vector]
data_sorted = [d for c,d in sorted(zip(projections, data), reverse=True)]

f, ax = subplots(figsize=(N*0.5,2))
plot(sorted(projections))
ylabel("PC" + str(vector+1))
show()
f, ax = subplots(1,N,figsize=(N*0.5,2))
for i in range(N):
    z = int(i / float(N) * len(data_sorted)) # even spacing over the pc
    img = data_default[z]
    ax[i].imshow(img.reshape((16,16)), interpolation='none', cmap = "Greys")
    ax[i].axis('off')
```

# reproduce the grid from ESL

```python
points = np.dot(data, VT.T)[:,:2]
pts = []
tree = KDTree(points) # fast search
N = 5 # pictures on one axis

minx = points.T[0].min()
miny = points.T[1].min()
maxx = points.T[0].max()
maxy = points.T[1].max()

xs = np.linspace(minx,maxx,N)
ys = np.linspace(miny,maxy,N)
grid = np.array([[(x,y) for x in xs] for y in ys])
f, ax = subplots(N, N, figsize=(N*1, N*1))

for i in range(N):
    for j in range(N):
        z = tree.query(grid[i,j])[1]
        img = data_default[z]
        pts.append(data[z])
        ax[i,j].imshow(img.reshape((16,16)), interpolation='none', cmap = "Greys")
        ax[i,j].axis('off')
```

```python
scatter(np.dot(U,S)[:,0], np.dot(U,S)[:,1], marker = ".") # data points
xlabel("PC1")
ylabel("PC2")
scatter(*zip(*np.dot(pts,VT.T[:,:2])), s = 75, marker = "*", c = 'r'  ) # chosen points
scatter(*np.vstack(grid).T, s = 75, marker = "+", c = 'r'  ) # grid points
```

# parallel analysis

```python
# generate random mats

n_repeats = 50
sim_eigs = np.zeros(shape = (n_repeats, data.shape[1]))

for i in xrange(n_repeats):
    m = np.random.normal(size = data.shape)
    sim_eigs[i] = np.square(svd(m)[1]) / m.shape[0]
```

```python
cent = 0 # or 0
if cent != 0:
    bias = np.percentile(sim_eigs, cent, axis = 0) - 1
else:
    bias = sim_eigs.mean(axis = 0) - 1 

evs = np.square(svd(data / data.std(axis = 0))[1]) / data.shape[0] # don't forget to set var = 1
adj_ev = evs - abs(bias) 
retain = adj_ev[adj_ev > 1].shape[0]

_ = plot(adj_ev, lw = 1, markersize = 2.5, marker = "o") #
_ = plot(np.ones(shape = adj_ev.shape)) # cutoff at 1
_ = title("Retain first " + str(retain) + " PCs after adjustment")
```

```python
names =  ["Moscow", "Novosibirsk", "St. Petersburg", "Dresden", "Warsaw", "Minsk", "Volgograd"]
D = [[0     ,2812.7,  631.9, 1654.5, 1148.2,  673.1, 906.4],
     [2812.7,     0, 3108.3, 4453.8, 3958.3, 3484.6, 2681.4],
     [631.9, 3108.3,      0, 1424.6, 1024.1,  686.2, 1536.4],
     [1654.5,4453.8,1424.6,       0,  517.4,  986.1, 2205],
     [1148.2,3958.3,1024.1,   517.4,      0,  475.3, 1702.5],
     [673.1,3484.6,686.2,     986.1,   475.3 ,    0, 1306.2],
     [906.4,2681.4,1536.4,     2205,     1702.5,  1306.2, 0]]
Dsq = np.square(np.array(D))
```

```python
H = np.eye(Dsq.shape[0]) - np.ones(shape = Dsq.shape) / Dsq.shape[0]
B = np.dot(H, np.dot(Dsq,H.T)) / -2
```

```python
eig, eiv = np.linalg.eig(B)
eig, eiv = zip(*sorted(zip(eig,eiv), reverse = True))
eig = np.array(eig) / np.array(eig).sum()
print "Eigenvalues: "+" ".join(map(str,eig))
print "small negative EVs occur due to the fact that cities' distances are defined on an elipsoid (Earth)"
plot(eig)
_ = title("Sorted eigenvalues of B")
```

```python
# remove elements close to zero
eivk = np.array(eiv)[:,eig > 1e-6]
eigk = eig[eig > 1e-6]
```

```python
# some coeffs to scale the result to the image
s = 1200
dx = 575
dy = 200
```

```python
xy = np.dot(eivk,np.diag(np.sqrt(eigk)))
xy = zip(*[(-s*i[0]+dx, s*i[1]+dy) for i in xy])
```
